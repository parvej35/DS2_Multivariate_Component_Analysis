```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

**Regression models** are used to describe relationships between variables by fitting a line to the observed data. Regression allows you to estimate how a dependent variable changes as the independent variable(s) change.

**Multiple Linear Regression (MLR)** is a regression model that estimates the relationship between a quantitative dependent variable and two or more independent variables using a straight line.

**Multiple Linear Regression** is used to estimate the relationship between two or more independent variables and one dependent variable.

**Simple regression** pertains to one dependent variable (y) and one independent variable **y=f(x)**

**Multiple regression** pertains to one dependent variable and multiple independent variables: 
**y=f(x1,x2,...,xn)**

**Multivariate regression** pertains to multiple dependent variables and multiple independent variables:
**y1,y2,...,ym = f(x1,x2,...,xn)**

Features consist of hourly average ambient variables
- Temperature (T) in the range 1.81°C and 37.11°C,
- Ambient Pressure (AP) in the range 992.89-1033.30 milibar,
- Relative Humidity (RH) in the range 25.56% to 100.16%
- Exhaust Vacuum (V) in teh range 25.36-81.56 cm Hg
- Net hourly electrical energy output (EP) 420.26-495.76 MW
The averages are taken from various sensors located around the plant that record the ambient variables every second. The variables are given without normalization.



### Load the dataset

```python
data_df = pd.read_csv("dataset_mlr.csv")
data_df.head()
data_df.info()
```

```python
#Plot the independent (AT, V, AP, RH) and dependent (PE) variables

sns.pairplot(data_df, x_vars=['AT', 'V','AP', 'RH'], y_vars='PE', height=7, aspect=0.7)
```

```python
sns.heatmap(data_df.corr(), annot=True)
```

### Linear Regression (Without data split): Just for experiment 

```python
#Linear Regression (Without data split): Just for experiment 
from sklearn.linear_model import LinearRegression

x = data_df[['AT', 'V','AP', 'RH']]
y = data_df.PE

ml = LinearRegression()
ml.fit(x, y)
print(ml.intercept_)
print(ml.coef_)

list(zip(['AT', 'V','AP', 'RH'], ml.coef_))
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

ml_pred = ml.predict(x[['AT', 'V','AP', 'RH']])

# 'y' is the actual value, 'ml_pred' is the predicted value
print("RMSE:", np.sqrt(mean_squared_error(y, ml_pred)))
print("R^2: ", r2_score(y, ml_pred))
```

### Define independent (X) and dependent (Y) variables

```python
# Dropping the column 'PE'.
# PE (Net hourly electrical energy output)is the dependent variable 
x = data_df.drop(['PE'], axis=1).values
x
```

```python
y = data_df['PE'].values #dependent variable
y
```

### Split the dataset in training and test set

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=0)
```

```python
# print(x_train)
# print(x_test)
# print(y_train)
# print(y_test)
```

### Train the model on the training set 

```python
from sklearn.linear_model import LinearRegression

ml = LinearRegression()
ml.fit(x_train, y_train)
```

### Predict the test set result

```python
y_predict = ml.predict(x_test)
print("Predicted values are:")
y_predict
```

```python
#Check the predicted values
print("Original value (PE) is : 463.26")
print("Predicted value is :")
#get the data from first row of the main data-set
ml.predict([[14.96, 41.76, 1024.07, 73.17]])
```

### Evaluate the model using r2_Score

```python
from sklearn.metrics import r2_score

print("Predicted ratio is :")
r2_score(y_test, y_predict)
```

### Plot the result

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plt.scatter(y_test, y_predict)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted")
```

### Predicted values

```python
pred_y_df = pd.DataFrame({"Actual Values":y_test, "Predicted Values":y_predict, "Difference":y_test-y_predict})
pred_y_df.head(20)
```
